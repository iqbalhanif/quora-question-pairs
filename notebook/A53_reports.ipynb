```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
hist_dict = pickle.load(open('../reports/04-05-2020_reports.pkl', 'rb'))
```

```python
pd.DataFrame(hist_dict)
```

## model 1

```python
hist_dict = pickle.load(open('../reports/model_5_v1.pkl', 'rb'))
```

```python
plt.plot(hist_dict["running_loss"], label = 'train')
plt.plot(hist_dict["running_loss_v"], label = 'val')
plt.title("loss model_5_v1")
plt.legend()
plt.savefig("loss.png")
```

```python
plt.plot(hist_dict["running_accu"], label = 'train')
plt.plot(hist_dict["running_accu_v"], label = 'val')
plt.title("accuracy model_5_v1")
plt.legend()
plt.ylim(50, 100)
plt.savefig('accuracy.png')
```
